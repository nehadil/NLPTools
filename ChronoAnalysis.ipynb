```python
import glob
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import plot, iplot
path = '/home/jeff/Documents/TAC_SRIE_2018_Task1_Training_Annotation/*.ann'
tacfiles = glob.glob(path)
tactrack = 0
chronotrack = 0
dfthread = {}
duthread = {}
tdthread = {}
tuthread={}
for file in tacfiles:
    tac = open(file, "r")
    name = file.split("/")[len(file.split("/"))-1]
    print(name)
    
    for line in tac:
        line = line.replace(";"," ")
        tactokens = line.split()
        if "DoseFrequency" in tactokens or "DoseDurationUnits" in tactokens or "TimeAtDose" in tactokens or "TimeUnits" in tactokens:
            chrono = open("/home/jeff/Documents/TACChrono-master/results/my_output"+"/"+name.split(".")[0]+"/"+name, "r")
            flag = False
            tactrack = tactrack +1
            for line2 in chrono:
                if flag == True and "type" in line2:
                    if "DoseFrequency" in tactokens:
                        if line2[8:-8] not in dfthread:
                            dfthread[line2[8:-8]] = 1
                        else:
                            dfthread[line2[8:-8]] = dfthread[line2[8:-8]]+1
                    elif "DoseDurationUnits" in tactokens:
                        if line2[8:-8] not in duthread:
                            duthread[line2[8:-8]] = 1
                        else:
                            duthread[line2[8:-8]] = duthread[line2[8:-8]]+1
                    elif "TimeAtDose" in tactokens:
                        if line2[8:-8] not in tdthread:
                            tdthread[line2[8:-8]] = 1
                        else:
                            tdthread[line2[8:-8]] = tdthread[line2[8:-8]]+1
                    else:
                        if line2[8:-8] not in tuthread:
                            tuthread[line2[8:-8]] = 1
                        else:
                            tuthread[line2[8:-8]] = tuthread[line2[8:-8]]+1
                    flag=False
               
                if "<span>" in line2 :
                    if line2[8:].split(",")[0]==tactokens[2] or line2[8:].split(",")[1][:-8]==tactokens[2]:
                        chronotrack = chronotrack+1
                        flag=True
          
        chrono.close()          
    tac.close()

    
print("tactotal:",tactrack)
print("chronototal:",chronotrack)
print(dfthread)
print(duthread)
print(tdthread)
print(tuthread)

dfx = []
dfy = []
dux = []
duy = []
tdx = []
tdy = []
tux = []
tuy = []

for thing in dfthread:
    dfx.append(thing)
    dfy.append(dfthread[thing])
for thing in duthread:
    dux.append(thing)
    duy.append(duthread[thing])
for thing in tdthread:
    tdx.append(thing)
    tdy.append(tdthread[thing])
for thing in tuthread:
    tux.append(thing)
    tuy.append(tuthread[thing])
dftrace = go.Bar(x=dfx,y=dfy,name="Dose Frequency")
dutrace = go.Bar(x=dux,y=duy,name="Dose Duration Units")
tdtrace = go.Bar(x=tdx,y=tdy,name="Time at Dose")
tutrace = go.Bar(x=tux,y=tux,name="Time Units")
layout = go.Layout(
    barmode='group'
)
fig = go.Figure(data = [dftrace, dutrace, tdtrace, tutrace], layout = layout)
plot(fig)
```
